```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Weather"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
print(soup.get_text())
```

```python
res = soup.title
```

```python
print(res.prettify())
```

```python
print(res.get_text())
```
